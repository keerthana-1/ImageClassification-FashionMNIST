#Random Forest Classification

#Initial Model

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load the Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Reshape the data into 1D arrays
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

# Split the data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Initialize and train a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Predict on the validation set
val_predictions = rf_classifier.predict(X_val)

# Calculate accuracy on the validation set
val_accuracy = accuracy_score(y_val, val_predictions)
print("Validation Accuracy:", val_accuracy)

# Evaluate on the test set
test_predictions = rf_classifier.predict(X_test)
test_accuracy = accuracy_score(y_test, test_predictions)
print("Test Accuracy:", test_accuracy)

# Generate and print classification report
print("Classification Report:")
print(classification_report(y_test, test_predictions))

# Show some random images from the dataset along with their labels
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    idx = np.random.randint(0, len(X_test))
    ax.imshow(X_test[idx].reshape(28, 28), cmap='gray')
    ax.set_title(f"True: {y_test[idx]}, Predicted: {test_predictions[idx]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

#With Dimensionality Reduction using PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Load the Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Reshape the data into 1D arrays
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

# Normalize the pixel values to range [0, 1]
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Split the data into train and validation sets
X_train_pca, X_val_pca, y_train, y_val = train_test_split(X_train_pca, y_train, test_size=0.2, random_state=42)

# Initialize and train a Random Forest classifier with tuned hyperparameters
rf_classifier = RandomForestClassifier(n_estimators=500, max_depth=20, min_samples_split=5, min_samples_leaf=2, random_state=42)
rf_classifier.fit(X_train_pca, y_train)

# Predict on the validation set
val_predictions = rf_classifier.predict(X_val_pca)

# Calculate accuracy on the validation set
val_accuracy = accuracy_score(y_val, val_predictions)
print("Validation Accuracy:", val_accuracy)

# Evaluate on the test set
test_predictions = rf_classifier.predict(X_test_pca)
test_accuracy = accuracy_score(y_test, test_predictions)
print("Test Accuracy:", test_accuracy)

# Generate and print classification report
print("Classification Report:")
print(classification_report(y_test, test_predictions))

# Show some random images from the dataset along with their labels
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    idx = np.random.randint(0, len(X_test))
    ax.imshow(X_test[idx].reshape(28, 28), cmap='gray')
    ax.set_title(f"True: {y_test[idx]}, Predicted: {test_predictions[idx]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

# Parameter Tuning with different set of values

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler

# Load the Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Reshape the data into 1D arrays
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

# Normalize the pixel values to range [0, 1]
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train a Random Forest classifier with manually chosen hyperparameters
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=20, min_samples_split=5, min_samples_leaf=2, random_state=42)
rf_classifier.fit(X_train_scaled, y_train)

# Predict on the test set
test_predictions = rf_classifier.predict(X_test_scaled)

# Calculate accuracy on the test set
test_accuracy = accuracy_score(y_test, test_predictions)
print("Test Accuracy:", test_accuracy)

# Generate and print classification report
print("Classification Report:")
print(classification_report(y_test, test_predictions))

# Show some random images from the dataset along with their labels
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    idx = np.random.randint(0, len(X_test))
    ax.imshow(X_test[idx].reshape(28, 28), cmap='gray')
    ax.set_title(f"True: {y_test[idx]}, Predicted: {test_predictions[idx]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Initialize and train a Random Forest classifier with reduced complexity
rf_classifier_simple = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42)
rf_classifier_simple.fit(X_train_scaled, y_train)

# Predict on the test set
test_predictions_simple = rf_classifier_simple.predict(X_test_scaled)

# Calculate accuracy on the test set
test_accuracy_simple = accuracy_score(y_test, test_predictions_simple)
print("Simple Random Forest Test Accuracy:", test_accuracy_simple)

# Generate and print classification report for the simplified Random Forest
print("Classification Report for Simple Random Forest:")
print(classification_report(y_test, test_predictions_simple))
```

#Trained model by obtaining feature importances

```python
# Train a Random Forest model
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_scaled, y_train)

# Get feature importances
feature_importances = rf_classifier.feature_importances_

# Sort feature importances in descending order
sorted_indices = np.argsort(feature_importances)[::-1]

# Select top features (e.g., top 200)
top_features_indices = sorted_indices[:200]
X_train_selected = X_train_scaled[:, top_features_indices]
X_test_selected = X_test_scaled[:, top_features_indices]

# Train a Random Forest model on the selected features
rf_classifier_selected = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier_selected.fit(X_train_selected, y_train)

# Predict on the test set
test_predictions_selected = rf_classifier_selected.predict(X_test_selected)

# Calculate accuracy on the test set
test_accuracy_selected = accuracy_score(y_test, test_predictions_selected)
print("Random Forest Test Accuracy (Selected Features):", test_accuracy_selected)

# Generate and print classification report for the Random Forest with selected features
print("Classification Report for Random Forest (Selected Features):")
print(classification_report(y_test, test_predictions_selected))
```
